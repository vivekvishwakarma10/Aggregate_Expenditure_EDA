```python
df = pd.read_csv("Aggregate_Expenditure.csv", encoding ='latin1')
```

```python
df
```

```python
rows,columns = df.shape
print("Rows: ",rows)
print("Columns: ",columns)
```

```python
for i in range(columns):
    print(str(i)+". :", df.iloc[:,i].dtype)
```

```python
df.isna().sum()
```

```python
df.replace({"":0},inplace=True)
```

```python
df
```

```python
for i in range(columns):
    print(str(i)+". :", df.iloc[:,i].dtype)
```

```python
lst = []
col = df.columns
print(col)
for i in col:
    if i!="State":
        lst.append(i)
print(lst)
```

```python
df[lst] = df[lst].apply(pd.to_numeric)
```

```python
for i in range(columns):
    print(str(i)+". :", df.iloc[:,i].dtype)
```

```python
plt.figure(figsize=(30,10))
sns.heatmap(df.corr(),annot=True,linewidth=1)
```

```python
plt.figure(figsize=(30,10))
sns.barplot(x='1980-81',y="State",data=df)
```

```python
for i in lst:
    plt.xlabel(i)
    plt.ylabel("State")
    plt.title(i+" vs State")
    plt.figure(figsize=(30,10))
    sns.barplot(x=i,y="State",data=df)
```
